```python
url = 'https://raw.githubusercontent.com/okHotel/ssd_project/master/data_set/istruzione_italia_66_14.csv'
df1 = pd.read_csv(url)
print(df1.head())
```

```python
df1 = df1[['anno', 'universita']]
print(df1.head())
```

```python
df1.sort_values(by='anno', inplace=True, ascending=True)
print(df1.head())
```

```python
fig = px.line(df1, x = 'anno', y = 'universita', title='Iscritti all\'universita\' anni 1966 - 2014 (in migliaia)')
fig.show()
```

```python
df1.insert(0, 'id', range(0,len(df1)))
print(df1.head())
```

```python
from sklearn.model_selection import train_test_split
X = df1[['id']].values.reshape(-1,1)
y = df1[['universita']].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVR
```

```python
# !!! Con kernel rbf scazza le previsioni.
# !!! Con kernel linear molto meglio !!!
regressor = SVR(kernel='linear')
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)

df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```
